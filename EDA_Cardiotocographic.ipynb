Outliers are extreme values that may skew the results or cause misleading conclusions.
In this step, boxplots are used to visualize the distribution of values for each feature and identify outliers.
These are shown as points outside the whiskers in boxplots.

```python
plt.figure(figsize=(15, 8))
df.boxplot(rot=90)
plt.title("Boxplot of all features")
plt.show()
```

```python
df.describe()
```

In this step, we visualize how each variable is distributed:

Histograms show how frequently values occur across intervals.

Boxplots reinforce the presence of outliers and provide a summary view of central tendency and spread.

These visualizations help detect skewness, peaks, and gaps in the data.

```python
df.hist(figsize=(16, 12), bins=30)
plt.suptitle("Histogram of All Features")
plt.show()

plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.xticks(rotation=90)
plt.show()
```

A correlation matrix shows the relationship (linear) between pairs of numerical variables.
The heatmap visualization color-codes the strength of correlation:

Values near 1 or -1 mean strong correlation.

Values near 0 mean weak or no correlation.

This helps in identifying which variables move together and could be useful in predictive modeling.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Two important visualizations are used here:

Pair Plot: Displays scatterplots between multiple pairs of variables to help identify linear/non-linear relationships.

Violin Plot: Compares the distribution of a variable (like LB) across different classes (like NSP). This helps understand how well a feature distinguishes between target categories.

```python
sns.pairplot(df.iloc[:, :6])
plt.show()

sns.violinplot(x="NSP", y="LB", data=df)
plt.title("Violin Plot of LB by NSP")
plt.show()
```

This step checks how many samples exist for each class of the target variable NSP (Normal = 1, Suspect = 2, Pathologic = 3).
A bar chart helps identify class imbalance.
This is important because an imbalanced dataset may bias the machine learning models toward the majority class.

```python
df["NSP"].value_counts().plot(kind='bar')
plt.title("Class Distribution of NSP")
plt.xlabel("NSP Class")
plt.ylabel("Count")
plt.show()
```
